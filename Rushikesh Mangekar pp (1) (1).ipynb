### Step 5-Creating the dummies for categorical variables

```python
df_dummies = pd.get_dummies(df[['Zone_Class', 'Road_Type', 'Alley', 'Property_Shape', 'LandContour',
                                'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1',
                                'Condition2', 'Dwelling_Type', 'HouseStyle', 'RoofStyle', 'RoofMatl',
                                'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond',
                                'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                                'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical',
                                'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType',
                                'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC',
                                'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']])
```

```python
df_dummies.head()
```

### Step 6- Concatinating the dummies and numerical variables.

```python
df_combined = pd.concat([df_numeric,df_dummies],axis=1)
```

```python
df_combined.info
```

### Step 7-Building the Model

```python
Y= df_combined['Property_Sale_Price']                       #Dependent Variable
X= df_combined.drop(columns='Property_Sale_Price',axis=1)   #Independent variable
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y,train_size=0.3, random_state=100)
```

### Step 8-Model fitting

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
# TRAIN DATASET EVALUATION

lm.fit(X_train,y_train)
y_train_pred=lm.predict(X_train)     # Prediction based on Fitted Model of lm
```

```python
#Find the coefficient of Determination
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score  



print('Mean squared Error for Regression',mean_squared_error(y_train,y_train_pred))
print('Root Mean squared Error for Regression',np.sqrt(mean_squared_error(y_train,y_train_pred)))
print('The coefficient of Determination is',r2_score(y_train,y_train_pred))
```

### Observed R^2 is 95.89%

```python
# TEST DATASET EVALUATION

lm.fit(X_test,y_test)
y_test_pred=lm.predict(X_test)     # Prediction based on Fitted Model of lm
```

```python
#Find the coefficient of Determination
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score  


print('Mean squared Error for Regression',mean_squared_error(y_test,y_test_pred))
print('Root Mean squared Error for Regression',np.sqrt(mean_squared_error(y_test,y_test_pred)))
print('The coefficient of Determination is',r2_score(y_test,y_test_pred))
```

### Observed R^2 is 93.99%

### Variance inflation factor
Variance inflation factor(VIF) for an explanatory variable is given 1/(1-R^2) Here, we take that particular X as response varible and all other explanatory varibles are indepent variables. So, we run a regerssion between one of those exlpanatory varibles with the remaining explanatory variables.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = df_numeric.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(df_numeric.values, i)
                          for i in range(len(df_numeric.columns))]
```

```python
vif_data[(vif_data['VIF']>5).sort_values(ascending=False)]
```
