# Step 3 - Splitting the data in Traning and Test set

using sklearn we split 70% of our data into training set and rest in train set.

Setting random state we give the same training and test everytime on running  the code

```python
# Putting feature Variables to x
x =LabourData_combined[['Age',  'Earnings_1974', 'Earnings_1975',
       'Race_NotBlack', 'Race_black', 'Hisp_NotHispanic', 'Hisp_hispanic',
       'MaritalStatus_Married', 'MaritalStatus_NotMarried',
       'Eduacation_HighSchool', 'Eduacation_Intermediate',
       'Eduacation_LessThanHighSchool', 'Eduacation_PostGraduate',
       'Eduacation_graduate']]
# Putting Response variables to Y
y= LabourData ['Earnings_1978']
```

```python
# random_state is the seed used by the random nuber generator. It can be any integer.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7, random_state = 100)
```

# Step 4- Performing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
#Representing Linearregression as lr(Creating Linearregression Object)
lm= LinearRegression()
```

```python
#fit the  the model to the tranning data
lm.fit (x_train,y_train)
```

# Step 5- Model Evaluation

```python
#print the intercept
print(lm.intercept_)
```

```python
# letr's see the coefficient 
coeff_df =pd.DataFrame (lm.coef_, x_test.columns,columns=['Coefficient'])
coeff_df
```

form the above results we may intern that if age incress by 1 unit 
it will affect Earnings_1978 by -105.198175 units


# Step 6 - Predictions

```python
#make predictions using model
y_pred = lm.predict(x_test)
```

```python
y_pred
```

## Step 7: Model Performance Metrics
### Coefficient of Determination(R Square)

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
```
